```python
import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt

df = pd.read_csv(r'https://media.githubusercontent.com/media/edwinjosechittilappilly/EDA-EV-ML-Project/main/datasets/GasPrice.csv')

array=df.to_numpy()

states=[]
GasPrice=[]
for element in array:
    states.append(element[0])
    GasPrice.append(element[1])
    
df2 = pd.read_csv(r'C:\Users\mclax\Documents\CS 5821 copy\Cost of Living.csv')

CofL=[]

array2=df2.to_numpy()
for element in array2:
    CofL.append(element[1])
    

df3 = pd.read_csv(r'C:\Users\mclax\Documents\CS 5821 copy\Cost of EV.csv')

CofE=[]

array3=df3.to_numpy()
for element in array3:
    CofE.append(element[1])

df4 = pd.read_csv(r'C:\Users\mclax\Documents\CS 5821 copy\GDP per Capita.csv')

GDPpCapita=[]

array4=df4.to_numpy()
for element in array4:
    GDPpCapita.append(element[1])


df5 = pd.read_csv(r'C:\Users\mclax\Documents\CS 5821 copy\Gas Prices.csv')

GasP=[]

array5=df5.to_numpy()
for element in array5:
    GasP.append(element[1])
    
    
plt.scatter(CofE,CofL)
plt.title('Cost of EV vs Cost of Living')
plt.xlabel('Cost of EV (USD)')
plt.ylabel('Cost of Living Index')
plt.show()
plt.bar(states,CofL)
plt.title('Cost of Living Index per State')
plt.xlabel('States')
plt.ylabel('Cost of Living Index')
plt.show()
plt.bar(states,CofE)
plt.title('Cost of EV per State')
plt.xlabel('States')
plt.ylabel('Cost of EV (USD)')
plt.show()
plt.bar(states,GDPpCapita)
plt.title('GDP per Capita per State')
plt.xlabel('States')
plt.ylabel('GDP per Capita (USD)')
plt.show()
plt.scatter(CofE,GDPpCapita)
plt.title('Cost of EV vs GDP per Capita')
plt.xlabel('Cost of EV (USD)')
plt.ylabel('GDP per Capita (USD)')
plt.show()
plt.scatter(CofL,GDPpCapita)
plt.title('Cost of Living vs GDP per Capita')
plt.xlabel('Cost of Living')
plt.ylabel('GDP per Capita (USD)')
plt.show()
plt.scatter(CofE,GasP)
plt.title('Cost of EV vs Gas Prices per State')
plt.xlabel('Cost of EV')
plt.ylabel('Gas Prices (USD/Gallon)')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
df6 = pd.read_csv(r'C:\Users\mclax\Documents\CS 5821 copy\FullData.csv')
X = df6[["GDP Per Capita","CofL","Avg Gas Price"]]
y=df6[["CofE"]]
reg = LinearRegression().fit(X, y)
print(reg.score(X, y))
print(reg.coef_)

est=sm.OLS(y,X)
est2=est.fit()
print(est2.summary())
```
